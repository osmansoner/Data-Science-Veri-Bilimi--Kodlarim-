```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
n = pd.read_csv("dünyam.csv")
n
n.dropna(how="any")
```

```python
sns.jointplot(n["Doğum oranı"],n["Ortalama yaş"],data=n)
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n)
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="hex",color="r")
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,xlim=(-2,6),ylim=(0,10),kind="reg",color="r",size=7)
```
