```python
n = pd.read_csv("dünyam.csv")
n
n.dropna(how="any",inplace=True) #NaN değerler silinmezse hata veriyor.
```

```python
sns.distplot(n["Doğum oranı"])
```

```python
sns.set(style="dark",palette="muted",font_scale=1.2) #palette seaborndan seç, style:arka plan, font_scale: yazı boyutu
sns.distplot(n["Doğum oranı"],bins=20,kde=False,color="y") #kde grafik üzerindeki çizgi,bins blok sayısı
plt.tight_layout()
```

```python
sns.set(style="darkgrid",palette="muted",font_scale=1.2) #palette seaborndan seç, style:arka plan, font_scale: yazı boyutu
sns.distplot(n["Yıllık değişim oranı"],hist=False,rug=True,color="r") #hist bloklar, rug tabandaki çizgiler,
plt.tight_layout()
```

```python
sns.set(style="darkgrid",palette="muted",font_scale=1.2) #palette seaborndan seç, style:arka plan, font_scale: yazı boyutu
sns.distplot(n["Ortalama yaş"],hist=False,kde_kws={"shade":True},color="r") #hist bloklar, rug tabandaki çizgiler,kdr_kws iç doldurma
plt.tight_layout()
```

```python
sns.set(style="whitegrid",palette="Blues",font_scale=1.5) #palette seaborndan seç, style:arka plan, font_scale: yazı boyutu
sns.distplot(n["Nüfüs"],color="m")
plt.tight_layout()
```

```python
sns.set(style="darkgrid",font_scale=1.5)
f, axes = plt.subplots(2,2,figsize=(15,10))

sns.distplot(n["Doğum oranı"],bins=20,color="y",ax=axes[0,0])
sns.distplot(n["Yıllık değişim oranı"],hist=False,rug=True,color="r",ax=axes[0,1])
sns.distplot(n["Ortalama yaş"],hist=False,color="g",kde_kws={"shade":True},ax=axes[1,0])
sns.distplot(n["Nüfüs"],color="m",ax=axes[1,1])

plt.tight_layout()
```
