```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = ["ABD","ÇİN","JAPONYA","ALMANYA","İNGİLTERE","FRANSA"]
y = [20410,14090,5170,4210,2940,2930] #milli gelir
z = [63039,9957,40649,51558,44162,44915] # kişi başı milli gelir
```

```python
f = plt.figure(figsize=(10,6)) #figsize tablo boyutunu ayarlar.
axes= f.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2,figsize=(13,5))

axes[0].plot(x,y,"b")
axes[0].set_title("MİLLİ GELİR")

axes[1].plot(x,z,"g")
axes[1].set_title("KİŞİ BAŞI GELİR")

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,"orange",linewidth=3,label="Milli Gelir")
ax.plot(x,z,"purple",linewidth=3,label="Kişi Başı Gelir")

ax.legend() #parantez içine "loc=0,1,2,3....,9,10  arasından bir sayı yazılarak konum belirlenebilir default olarak 0'dır.
```
