```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures as pf
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from sklearn import linear_model
```

```python
data = pd.read_csv("linear.csv")
```

```python
x = data["metrekare"]
y = data["fiyat"]
```

```python
x = np.array(x)
y = np.array(y)
```

```python
a,b,c,d = np.polyfit(x,y,3)
```

```python
z = np.arange(150)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(z,a*(z**3)+b*(z**2)+c*(z**1)+d)
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(z,a*(z**3)+b*(z**2)+c*(z**1)+d)
plt.show()
```

```python
for i in range(int(len(y))):
    hatakaresi =+ ((y[i])-(a*(x[i])**3+b*(x[i])**2+c*(x[i])**1+d))**2
```

```python
print("Polinomsal hata:",hatakaresi)
```
