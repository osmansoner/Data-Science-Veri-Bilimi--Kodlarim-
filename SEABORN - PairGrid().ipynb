```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
e = pd.read_csv("50ülke.csv")
e
```

```python
n = pd.read_csv("dünyam.csv")
n.dropna(how="any",inplace=True)
n
```

```python
m = sns.PairGrid(n)
m
```

```python
k = sns.PairGrid(n)
k.map_diag(sns.distplot)
k.map_upper(plt.scatter)
k.map_lower(sns.kdeplot)
```

# FacetGrid()

```python
s = pd.read_csv("evli.csv")
```

```python
ss = sns.FacetGrid(data=s,col="ay",row="evli")
ss.map(sns.distplot,"gelir")
```
