```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = ["ABD","ÇİN","JAPONYA","ALMANYA","İNGİLTERE","FRANSA"]
y = [20410,14090,5170,4210,2940,2930] #milli gelir
z = [63039,9957,40649,51558,44162,44915] # kişi başı milli gelir
kısax = ["ABD","ÇİN","JPY","ALM","İNG","FRA"]
```

```python
plt.plot(x,y,"y") #y yellow yani renk belirtir.
plt.xlabel("ÜLKELER")
plt.ylabel("MİLLİ GELİR")
plt.title("ÜLKELERE GÖRE MİLLİ GELİR")
plt.show() #jupyter notebook da yazmamıza gerek yok aslında ama diğerlerinde gerekli
```

```python
plt.subplot(1,2,1) #1 satır 2 sütunlu grafik oluştur ve 1.grafik için şunlar;
plt.plot(kısax,y,"r--")
plt.subplot(1,2,2) # bu da 2.grafik
plt.plot(kısax,z,"g*-")
plt.tight_layout()
```

```python
f = plt.figure()
axes = f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
f = plt.figure()
axes1 = f.add_axes([0.1,0.1,0.9,0.9])
axes2 = f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y,"g")
axes2.plot(kısax,z,"y")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(kısax,y,"b")
axes[0].set_title("MİLLİ GELİR")
axes[1].plot(kısax,z,"r")
axes[1].set_title("KİŞİ BAŞI GELİR")
plt.tight_layout()
```
