marker yuvarlk olan ancak + vs herhangi bir şey olabilir

alpha saydamlık ayarı 0-1 arası değer alır 1'e yaklaştıkça mat olur.

```python
f = plt.figure()
axes = f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="orange",linewidth=2,linestyle="-.",marker="o",markersize=14,markerfacecolor="green",markeredgewidth=3,markeredgecolor="red",alpha=0.8)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```
